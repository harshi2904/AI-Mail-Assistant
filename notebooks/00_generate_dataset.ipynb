```python
import pandas as pd
import random

# Define categories and sample email templates
categories = {
    "IT Support": [
        "My laptop is not connecting to the Wi-Fi network.",
        "VPN connection keeps dropping frequently.",
        "Outlook is not syncing my emails.",
        "Please reset my system password.",
        "The printer in the 3rd floor office is not working."
    ],
    "HR": [
        "Please share my salary slip for last month.",
        "I want to apply for sick leave for two days.",
        "When will the appraisal process start?",
        "Can you update my bank details in the HR portal?",
        "I need the joining letter for my background check."
    ],
    "Finance": [
        "My reimbursement request has not been approved yet.",
        "Payment to vendor ABC is still pending.",
        "Please provide invoice for last month’s purchase.",
        "There’s an error in my tax deduction for this month.",
        "Can you share the expense policy document?"
    ],
    "Admin": [
        "The AC in meeting room A is not cooling properly.",
        "Need a new ID card, mine is damaged.",
        "Please book a meeting room for tomorrow.",
        "The projector is not functioning.",
        "Can you arrange stationery for the new interns?"
    ],
    "General": [
        "Good morning team, please find attached the weekly report.",
        "Let’s plan the team outing next week.",
        "Thank you for your quick response!",
        "Please review the attached document and share feedback.",
        "Can we schedule a call to discuss the new policy?"
    ]
}

```

```python
# Generate dataset
data = []
for category, messages in categories.items():
    for i in range(1, 40):  # generate ~40 per category
        body = random.choice(messages)
        subject = f"{category} Query #{i}"
        data.append({
            "id": len(data)+1,
            "subject": subject,
            "body": body,
            "category": category
        })

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('../data/raw_emails.csv', index=False)

print("✅ Synthetic dataset created and saved to data/raw_emails.csv")
df.sample(5)
```
