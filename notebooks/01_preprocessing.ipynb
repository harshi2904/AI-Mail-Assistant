```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
# Download NLTK resources (only first time)
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Load dataset
df = pd.read_csv('../data/raw_emails.csv')

# Initialize preprocessing tools
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(text):
    # Lowercase
    text = str(text).lower()
    # Remove emails, URLs, special characters, digits
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize and remove stopwords
    words = [lemmatizer.lemmatize(w) for w in text.split() if w not in stop_words]
    return " ".join(words)
```

```python

# Apply cleaning
df['clean_body'] = df['body'].apply(clean_text)
```

```python
# Combine subject + body for final feature
df['text'] = df['subject'].fillna('') + " " + df['clean_body']
```

```python
# Save processed data
df[['id', 'text', 'category']].to_csv('../data/processed_emails.csv', index=False)

print("✅ Preprocessing complete. Saved to data/processed_emails.csv")
df.sample(5)
```
