```python
# 03_llm_summary_hf.ipynb

import pandas as pd
from transformers import pipeline
from tqdm import tqdm
```

```python
# Load summarization model from Hugging Face
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
# ✅ Define summarization function
def summarize_email_local(text):
    """Summarize email text in one or two lines using BART."""
    try:
        summary = summarizer(text, max_length=40, min_length=10, do_sample=False)
        return summary[0]['summary_text']
    except Exception as e:
        return f"Error: {e}"
```

```python
# ✅ Define polite AI reply generator (template-based)
def generate_reply_local(text, category):
    """Generate a polite reply template for each category."""
    templates = {
        "IT Support": "Hi, thanks for reaching out. We’ll look into your IT issue right away and keep you posted.",
        "HR": "Hello, we’ve received your HR-related request and will process it shortly. Thank you for your patience.",
        "Finance": "Hi, your finance query has been received. Our team will review and follow up soon.",
        "Admin": "Hello, your admin request has been logged. We’ll ensure it’s resolved promptly.",
        "General": "Hi, thank you for your message. We appreciate your update and will get back to you soon."
    }
    # Default fallback
    return templates.get(category, "Thank you for reaching out. We’ll get back to you soon.")
```

```python
# ✅ Load processed dataset
df = pd.read_csv('../data/processed_emails.csv')
```

```python

# Sample 10 rows for testing (to avoid slow long runs)
df_sample = df.sample(10, random_state=42).copy()
```

```python
# Generate summaries and replies
summaries, replies = [], []

for _, row in tqdm(df_sample.iterrows(), total=len(df_sample)):
    text = row['text']
    category = row['category']
    summaries.append(summarize_email_local(text))
    replies.append(generate_reply_local(text, category))

df_sample['summary'] = summaries
df_sample['ai_reply'] = replies
```

```python


# Save results
df_sample.to_csv('../data/emails_with_ai.csv', index=False)

print("✅ Local summaries & replies generated successfully!")
df_sample[['text', 'category', 'summary', 'ai_reply']]
```
